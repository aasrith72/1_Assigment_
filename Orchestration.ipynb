```python
from transformers import pipeline
```

```python
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
```

```python
flan_pipeline=pipeline("text2text-generation", model="google/flan-t5-small")
```

```python
content="""
LangChain is a framework for building applications powered by language models.
It helps developers connect LLMs tools, memory, chains, and data retrieval.
"""
```

```python
question="What is LangChain?"
context_answer = qa_pipeline(question=question, context=content)["answer"]
```

```python
final_input=f"Q:{question}\nA (from context):{context_answer}\nNow summarize it clearly."
response=flan_pipeline(final_input, max_new_tokens=128)
```

```python
print("\n Final Answer:", response[0]['generated_text'])
```
